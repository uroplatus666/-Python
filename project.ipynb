На часть вопросов (символ и слава Кургана, посещаемые города) респонденты давали сразу несколько ответов. Напишите функцию, которая будет получать на вход столбец датафрейма (например, символ), а выводить **словарь**, где **ключ** - вариант ответа, а **значение** - количество опрошенных, которые так ответили.

```python
dictt = {}
def to_dict(text):
    if not pd.isna(text):
        text=str(text)
        for i in text.split(', '):
            if i in dictt:
                dictt[i] += 1
            else:
                dictt[i] = 1
    return dictt
```

Выведите 3 символа города, которые называли чаще всего.

```python
dictt = {}
symbology = data['Символ города'].apply(to_dict)[0]
```

```python
for symbol in sorted(symbology.items(), key = lambda i: i[1], reverse = True)[:3]:
    print(symbol[0])
```
